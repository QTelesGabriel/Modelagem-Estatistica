```python
import pandas as pd
from scipy import stats
from seaborn import distplot

dados = pd.read_csv('DadosIDEB1.csv')
dados.head()
```

```python
df = pd.DataFrame(dados)
df2 = df.drop(labels=['Codigo', 'UF', 'Municipio', 'Rede'], axis=1)
```

```python
print("Média e Desvio padrão")
df.describe().loc[['mean', 'std'], 'Ap_Media':'IDEB']
```

```python
print("Moda")
df.mode()
```

```python
df2.boxplot()
```

```python
df2.hist()
```

```python
z_score = stats.zscore(df[['Ap_Media', 'Nota_SAEB_Mat', 'Nota_SAEB_ LP', 'IDEB']])
print(z_score)
distplot(z_score)
```
