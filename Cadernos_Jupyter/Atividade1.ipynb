```python
import numpy as np
from seaborn import distplot
from random import randint
```

```python
idades = []

tamanho_entre_0_12 = 12500
tamanho_entre_13_35 = 25000
tamanho_entre_35_55 = 7500
tamanho_maior_56 = 5000
```

```python
for i in range(tamanho_entre_0_12):
    resultado = randint(0, 12)
    idades.append(resultado)

for i in range(tamanho_entre_13_35):
    resultado = randint(13, 35)
    idades.append(resultado)
    
for i in range(tamanho_entre_35_55):
    resultado = randint(35, 55)
    idades.append(resultado)
    
for i in range(tamanho_maior_56):
    resultado = randint(56, 100)
    idades.append(resultado)
```

```python
distplot(idades)
```

```python
# Média e desvio padrão

media = np.mean(idades)
desvio_padrao = np.std(idades)

media, desvio_padrao
```

```python
# 100 amostras de 100 pessoas

amostras = 100
medias = []

for i in range(100):
    amostra = np.random.choice(idades, size = amostras)
    media = amostra.mean()
    medias.append(media)
    
distplot(medias)
```

```python
# Z-score

amostras = 100

media_amostras = []
z_score = []

for i in range(100):
    amostra = np.random.choice(idades, size = amostras)
    media_2 = np.mean(amostra)
    desvio_padrao_2 = np.std(amostra)
    for x in amostra:
        z_score.append((x - media_2) / desvio_padrao_2)

distplot(z_score)
```
