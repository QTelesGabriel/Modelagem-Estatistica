```python
esperado = 0
for evento, probabilidade in zip(eventos_dado, probabilidades_dado):
    print(evento)
    print(probabilidade)
    esperado = esperado + evento * probabilidade
    print(esperado)
```

```python
# Funções
def calcula_esperado(eventos, probabilidades):
    """
    Função para calcular o valor esperado de uma variável a partir dos eventos e sua probabilidades.
    RETURNS: float
    """
    esperado = 0
    for evento, probabilidade in zip(eventos_dado, probabilidades_dado):
        esperado = esperado + evento * probabilidade
    return esperado

esperado = calcula_esperado(eventos_dado, probabilidades_dado)
print(esperado)
```

```python
# Probabilidade Condicional com a Fórmula de Bayes
p_a = 0.1    # Probabilidade ou frequÇencia de ocorrência da gripe
p_b_dado_a = 0.75    # Probabilidade ou frequência de ocorrência da tosse dado que a pessoa está gripada
p_b = 0.75*0.1 + 0.25*0.9    # Probabilidade frequência de ocorrência da tosse

# Quando a pessoa está ou não está gripada
p_a_dado_b = p_b_dado_a * p_a/p_b    # A probabilidade de se estar com gripe(a), dado que se está com tosse(b)
print(f"A probabilidade de se estar com gripe(a), dado que se está com tosse(b) é {p_a_dado_b}.")
```
