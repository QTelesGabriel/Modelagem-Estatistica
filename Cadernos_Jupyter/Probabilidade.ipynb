```python
# Qual é a probabilidade da fumaça que estou vendo ser de um incêndio?
p_incendio = 0.01    # 1% de probabilidade de encontrar um incêndio
p_fumaca = 0.1    # 10% de probabilidade de encontrar fumaça
p_fumaca_dado_incendio = 0.9    # 90% de chance de um incêndio gerar fumaça
```

```python
p_incendio_dado_fumaca = p_incendio * p_fumaca_dado_incendio / p_fumaca
p_incendio_dado_fumaca
```

```python
# Se está nublado,, qual é a probabilidade que chova?
p_nublado_dado_chuva = 0.5    # 50% de chance de ter estado nublado dado que choveu
p_nublado = 0.4    # 40% de chance de amanhecer nublado
p_chuva = 0.1    # 10% de probabilidade de chuva nessa época do ano
```

```python
p_chuva_dado_nublado = p_chuva * p_nublado_dado_chuva / p_nublado
p_chuva_dado_nublado
```

```python
tamanho_amostra = 100
n_amostras = 1000
medias_amostras = []
for i in range(n_amostras):
    amostra = np.random.choice(x, size = tamanho_amostra, replace = False)
    media_amostra = amostra.mean()
    medias_amostras.append(media_amostra)
```
