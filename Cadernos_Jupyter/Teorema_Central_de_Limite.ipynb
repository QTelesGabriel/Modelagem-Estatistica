```python
from matplotlib import pyplot as plt
from random import randint, choice
import numpy as np
import pandas as pd
```

```python
jogadas = []

for i in range(10000):
    resultado = np.random.randint(1, 6)
    jogadas.append(resultado)
len(jogadas)
```

```python
# 10 amostras e 10 jogadas

import seaborn as sns
n_amostras = 10
t_amostra = 10
medias = []
for i in range(n_amostras):
    amostra = np.random.choice(jogadas, size = t_amostra)
    media = amostra.mean()
    medias.append(media)
    
sns.distplot(medias)
```
